```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import wrangle
```

```python
df = wrangle.prepare_TWINS()
df.head()
```

```python
# resample data to daily view
daily_sample = df[["twins_2L_TEMP","twins_air_TEMP"]].resample('D').mean()
df[["twins_2L_TEMP","twins_air_TEMP"]].resample('D').mean().plot();
```

```python
# Define the size each dataframe needs to hold
# So train will hold the first half of the data, validate the next 30%, and test will be the last of what's left
train_size = int(len(df) * .5)
validate_size = int(len(df) * .3)
# Calculating test based on train and validate
test_size = int(len(df) - train_size - validate_size)
# Defining where validate will stop in the index
validate_end_index = train_size + validate_size

# Now to actually split into train, validation, test
train = df[: train_size]
validate = df[train_size : validate_end_index]
test = df[validate_end_index : ]
```

```python
# Does validate start after train without gaps or overlap?
pd.concat([train.tail(1), validate.head(1)])
pd.concat([validate.tail(1), test.head(1)])
```

It doesn't look like they do, but there is in fact a time gap here. So the missing 4 seconds of data is appropriate. 

```python
# resampled daily data:
x = train.resample('D').twins_2L_TEMP.mean()
y = validate.resample('D').twins_2L_TEMP.mean()
z = test.resample('D').twins_2L_TEMP.mean()

x.head()
```

```python
# Visualize train
x.plot();
```

```python
# Visualize validate
y.plot();
```

```python
# Visualize the split data:
train.resample('D').twins_2L_TEMP.mean().plot();
```

```python
# Let's visualize the train, validate, test
for col in train.columns:
    plt.figure(figsize=(15,7))
    plt.plot(train[col])
    plt.plot(validate[col])
    plt.plot(test[col])
    plt.ylabel(col)
    plt.title(col)
    plt.show()
```
